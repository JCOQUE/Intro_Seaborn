## <a id = 'vs'>Matplotlib vs Seaborn</a>
Which one to use?

- This depends on your goals, of course. 

<br>

I've read that you should use *Matplotlib* when you want full control in what you are plotting. On the other hand, if you just want to make a quick basic plot, go for *Seaborn*.

In my opinion, this is not quite true. *Matplotlib* and *Seabron* have their own characteristics. *Matplotlib* has the [plt.subplot_mosaic](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot_mosaic.html) method that might be what we want. *Seaborn* has some plots, like [pairplot()](https://seaborn.pydata.org/generated/seaborn.pairplot.html) that *Matplotlib* doesn't have. 

In general, I believe that one should use Seaborn. Apparently, with *Matplotlib* you can configure and customize more your plots; but since *Seaborn* is built on top of *Matplotlib*, you can access to these low level functions as we'll see in the last notebook of this *Seaborn tutorial*. 

The drawback of *Seaborn* is that it is harder to learn. People will tell you that it is easier, but in my opinion, that's not true. It's easier for simple plots. But as we'll see, *Seaborn*, in order to make plotting *easier* has more attributes, many methods and many ways to plot - and some of them very similar which ends up being a bit confusing. 

Let's start.

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## <a id = 'intro'>Introduction</a>

Quick notes:

In this specific notebook, we'll see different plot functions that Seaborn has. If you do not understand how they work, some parameter or anything else don't worry. We'll introduce most of these functions (the most important ones) deeper in the following notebooks.

My goal in this notebook is for you to understand what a **figure level plot** is vs an **Axis level plot**. Also it is important for you to understand what I call the *'functions tree'*. You'll see that every **Axis level plot function** is the *child* of a **figure level plot function**.

>**Note**: From now on, to shorten these words a little bit, I am going to be referring to *figure-level functions* as `flf` and to *axis-level functions* as `alf`.

### <a id = 'diff'>What's the difference?</a>
They do have some different behavior. The main differences / characteristics are
- The `flf` plots `alf` functions. I mean, they are not a plot on their own. The `alf` function is passed through the `kind` attribute every `flf` has.
- The `flf` allows one to plot subplots, cotaining each one of them a `alf`. 
- The `flf` have their legend placed outside the plot, while the legend of the `alf` is placed inside.
- The `flf` signature is meant for figure configuration, although it accepts any child signature inside of its own signature. 

>**Note**: A function's signature is the parameters it receives.

Let's see the `flf` and `alf` in action to understand better the concept.

```python
# using the tips dataset
penguins = sns.load_dataset("penguins")
penguins
```

```python
# relplot figure-level function
sns.displot(data = penguins, x = 'island', hue = 'species',
           height = 3, aspect = 1.5, kind = 'hist', col = 'sex', binwidth = 1)
plt.show()
```

From the above plot, notice that:
- The legend is outside
- The `displot() afl` is plotting the `histplot() alf` passed in the attribute `kind`.
- We were able to make a subplot because we used a `flf`.
- We passed to the `flf` a `alf`'s attribute (`binwidth`, attribute of `histplot()`).


```python
# But, we could plot directly an histplot like we wanted
sns.histplot(data = penguins, x = 'species', hue = 'island')
```

>**Note**: The plots are meaningless. They are there just to show basic *Seaborn* characteristics that you must understand.

So... if I can plot a `histplot()` or a `scatterplot()` and any other `alf` both ways (using `flf` or with the `alf` directly), which one should you use? Does it matter?

#### This is what Seaborn says about using figure-level functions
> *The tutorial documentation mostly uses the figure-level functions, because they produce slightly cleaner plots, and we generally recommend their use for most applications.*

The only situation where the *Seaborn documentation* recommends using `alf` is when you want to plot subplots, and each subplot you want is a different `alf`.

**But, you said before that with `alf` you couldn't plot subplots...**
- You're right, but ironically, you can plot subplots with `alf` using matplotlib. And this way **is the only way to plot different plots in a subplot matrix**. 

We'll see this in more detail in the next notebook. With `flf` you can plot different plots in a subplot but with less flexibility as we'll see with [Pair Grid()](https://seaborn.pydata.org/generated/seaborn.PairGrid.html).


My main goal is that you get the idea that you can have (almost) the same graph by using `flf` and `alf`. But, in most cases, it's better to use a `flf` since you have the `flf` attributes + the `alf` attributes to plot, and you can make subplots (at least, much easier).

### <a id = 'hss'>Hue, size and style</a>
The Hue, size and style attributes are a `flf` and a `alf` attributes. 
- `hue` is used as a groupby of a variable.
- `size` marks with different sizes the different values of the associated variable.
- `style` marks with different styles the different values of te associated variable.

Also, the reason I explain these attributes together is because they allow one to plot more variables in a plot. 
But with these attributes you can plot up to three more variables (5 in total, x_axis, y_axis +3, one per attribute) in one plot (`alf`). 

Another way to approach this (plotting multiple variables), is by making subplots (`flf`), and plotting different variables in different rows and cols. Let's see both approaches with examples. They're going to make this clearer. 

```python
tips = sns.load_dataset("tips")
tips.head(7)
```

```python
# plotting two variables, x and y in a single plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
#plotting 3 variables (with hue) in a single plot. Hue is 
# going to do a groupby to the variable associated with it
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")

# accessing matplotlib's axis. Using matplotlib's API
g.ax.text(5, 9, 'Note that now we are\nusing a flf to plot\nthe same scatter plot', 
         va='center', bbox={'facecolor': 'orange', 'alpha': 0.5, 'pad': 10})
g.ax.text(0.6, 0.85, 'Note that we can access\nlow level matplotlib\nfunctionality with Seaborn\nthat is why I can access\nthe text method.', 
         va = 'center', bbox={'facecolor': 'orange', 'alpha': 0.5, 'pad': 10}, transform = g.ax.transAxes)
plt.plot()# this for not showing the text in the output
```

```python
#plotting 4 variables with hue and style
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time"
)
```

```python
# plotting 5 variables with hue, style and size
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="smoker", 
    style="time", size = 'size', sizes = (15, 100)
)
```

We can agree that it might get to a point where subplots are much better. I just wanted to show you that you can plot up to 5 variables in a single plot.

### <a id = 'colrow'>col and row</a>
To plot multiple variables with sublpots, we can use `hue`, `style`, `size` + `col` and `row`. Needlees to say that this way is much better in case we wanted to plot multiple variables

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
             col = 'time', row = 'smoker', height = 3)
```

Also you can target `hue` and `style` to the same variable. It will plot 3 variables (not 4) but it is recommended to use them both together. This is because when using `hue`, `hue` differenciates each class by color.But, what if someone prints our graph in black-and-white paper? That's why using `hue` along with `style` (or size) targetting the same variable is a good practice because it makes our graphs more accessible.

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', style = 'smoker')
```

### <a id = 'order'>row_order, col_order,  hue_order, style_order, size_order</a>

#### row_order & col_order

In the subplot we plotted a few cells above, our row variable was `smoker` and our col variable was `time`. They were plotted in a certain order:
- **First row**: *smoker = Yes*, **Second row**: *smoker = No*
- **First col**: *time = Lunch*, **Second col**: *time = Lunch*

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
             col = 'time', row = 'smoker', height = 3)
```

We can change this order with `row_order` and `col_order` if we wanted to. 

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
            col = 'time', row = 'smoker', row_order = ['No', 'Yes'],
            col_order = ['Dinner', 'Lunch'],height = 3)
```

#### hue_order, style_order, size_order

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="sex", style="time",
    size = 'smoker', sizes = [10,25], height = 5)
```

```python
#chaging the order
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="sex", style="time",
    size = 'smoker', sizes = [10,25], hue_order = ['Female', 'Male'],
    size_order = ['No', 'Yes'], style_order = ['Dinner',  'Lunch']
    
)
```

>**Note**: The order with `hue` only can be changed if the variable is categorical. This is because `hue`  draws *different colors* depending on whether the associated  variable is **numeric or categorical**.

## <a id = 'wrap'>col_wrap</a>

Sometimes, we have so many different *levels* of a variable (a boolean variable has two levels: True and False), you can plot the variable along the column but 'wrap' the levels so that they span in multiple rows,

```python
attend = sns.load_dataset("attention").query("subject <= 12")
attend.head(6)
```

```python
num_subjects = max(attend.subject.value_counts().index)
print(f'The levels of the variable subject are {num_subjects}')
```

```python
# col_wrap = 4. Each row will have 4 levels of the subject variable
g = sns.FacetGrid(attend, col="subject", col_wrap=4, height=2, ylim=(0, 10))
g.map(sns.pointplot, "solutions", "score", order=[1, 2, 3], color=".3", errorbar=None)
g.figure.text(0.1, 1.06, 'g.figure.suptitle', bbox={'facecolor': 'orange', 'alpha': 0.5, 'pad': 10})
g.figure.suptitle('flf: FacetGrid        alf: pointplot', y = 1.1)
```

<br><br>

**Notice that again, we access to the *Matplotlib* API using** `g.figure.suptitle()`.

<br><br>

### <a id = 'ha'>Height and Aspect</a>
Seaborn has some peculiarities when it comes to assgning height and width to the plot.

- First, when you assign `height` and `width` in Matplotlib `figsize = (6,6)`, you are declaring the **figure size**. In *Seaborn*, when you declare `height` and `width`, you are referring to **each subplot size**.

- Second, these parameters, `height` and `aspect`, parameterize the size slightly differently than the `width`, `height `parameterization in matplotlib (using the seaborn parameters, `width = height * aspect`).

<br>

So the `aspect` means the **width's ratio** compared to the `height`. 

- An `aspect = 0.5` means that the subplot will have half `width` lenght respect to the `height` length.
- An `aspect = 2` means that the subplot will have twice the length of the `height`. 

```python
#this is to visualize the height and width better.
sns.set_theme()

g = sns.FacetGrid(penguins, col="sex", height=3.5, aspect=.75)
```

### <a id = 'pal'>palette (color)</a>
With `palette` you can set different color styles to your plot. You can pass many different string values, each one of them represeting a different color scale.

>**Trick**: In order to know all possible string values that can be  passed, pass to `palette` an empty string. In the error message you'll see all different and possible options.

```python
# error on purpose
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
             col = 'time', row = 'smoker', height = 3, palette = '')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
             col = 'time', row = 'smoker', height = 3, palette = 'rocket')
```

Another way to pass colors to `palette` is with the [cubehelix_palette](https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette).

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size',
             col = 'time', row = 'smoker', height = 3, palette = 'ch:r=-.5,l=.75')
```

### <a id = 'fk'>facet_kws</a>
`facet_kws` is an attribute that `relplot()`, `distplot()` and `catplot()`have. 

I will explain it more in detail in the next notebook, but `FacetGrid()` is like the father of `relplot()`, `distplot()` and `catplot()` (is the father of the fathers). It has some attributes (**very generic figure attributes**) that the other `flf` don't (not many). In order to pass to `relplot()`, `distplot()` and `catplot()` some attributes of the `FacetGrid()` class, we pass pass them with the `facet_kws` attribute.

>**Note**: Not sure if in newer *Seaborn* versions, the `facet_kws` content can be pass as normal attributes and *Seaborn* manages it behing scenes.

#### Sharex, Sharey, xlim, ylim.

They work in the same way as in Matplotlib. The only difference is that `sharex` and `sharey` **default value** is equal to `True`.

>**Note**: Notice that the `xlim` and `ylim` are defined in the **figure-level function** and not in the **axis-level function**.

```python
sns.relplot(data = penguins, x="bill_length_mm", y = 'bill_depth_mm', col="species", row="sex", 
            height = 2.7, facet_kws={'sharey': False, 'sharex': True})
```

You can also determine whether only `rows` or `cols` share either *x_axis* or *y_axis*.

```python
sns.relplot(data = penguins, x="bill_length_mm", y = 'bill_depth_mm', col="species", row="sex", 
            height = 2.7, facet_kws={'sharey': 'row', 'sharex': 'col'})
```

```python
sns.relplot(data = penguins, x="bill_length_mm", y = 'bill_depth_mm', col="species", row="sex", 
            height = 2.7, facet_kws={'xlim': [40, 50], 'ylim': [15, 20]})
```

#### legend_out

`Boolean` attribute to determine the legend placement. `True` by default.

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", height = 4)
```

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", height = 4,
                facet_kws={'legend_out': False})
```

#### margin_titles
`Boolean` attributes that determines where to place the **subplot's title**. `False` by default.

Only is necessary when we have `col` and `row` attributes set to some variables.

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col = 'sex', height = 4)
```

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col = 'sex', row = 'time', 
                height = 4, facet_kws={'margin_titles': True})
```

#### despine
Works the same way `ax.spine` does in *Matplotlib*. It  controls the figure's spines (i.e. borders) and you can make them dissapear or move them creating some spaces between *x_axis* and *y_axis*.

It can be used in two different ways (for the same purpose).
- as a `boolean` attribute. If set to True, it removes right and top spines. 
    - `True` by default.

```python
#removes set_theme
sns.reset_defaults()
#I need to add again this to plot correctly
%matplotlib inline

sns.relplot(data = tips, x= 'total_bill', y = 'tip', hue = 'smoker', height = 3)
```

```python
sns.relplot(data = tips, x= 'total_bill', y = 'tip', hue = 'smoker', 
            height = 3, facet_kws={'despine': False})
```

Second way of using **despine**.
- with `sns.despine`. It allows the user much more  flexibility.<br><br>

    - `seaborn.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)`.

>**Note**: `True` means remove.

```python
sns.relplot(data = tips, x= 'total_bill', y = 'tip', hue = 'smoker', 
            height =4)
sns.despine(bottom = True, left = True)
```

```python
#creating some space between x_axis and y_axis
sns.relplot(data = tips, x= 'total_bill', y = 'tip', hue = 'smoker', 
            height =4)
sns.despine(trim = True)
```

#### dropna

- Might be useful when using `displot` because the `NaN` values might affect the distribution or some attribute like `stats = percentage`.
