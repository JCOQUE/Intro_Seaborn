We've seen so far the `flf` (figure-level functions) and `alf`(axis-level functions). And that `FacetGrid` (a `flf`) was like the *father* of the next `flf`s: `relplot()`, `displot()` and `catplot()`. We can see the *function tree* in the next image:

In this notebook we are going to see is more depth the `relplot` family. 

It is not a coincidence that the `flf` are separated in that way. Each `flf` family is meant for a specific purpose, different from the rest of the families.


- `relplot` *family* is meant for **visualizing statistical relationships**. This can be done with the `alf`: `scatterplot` and `lineplot`.


- `displot` *family* is meant for **visualizing distributions of data**. This can be done with the `alf`: `histplot`, `kdeplot`, `ecdfplot` and `rugplot`. 


- `catplot` *family* is meant for **visualizing categorical data**. This can be done with the `alf`: `stripplot`, `swarmplot`, `boxplot`, `violinplot`, `pointplot` and `barplot`. 

<br><br>

In this notebook we'll be focusing on the `relplot` family.

## <a id='rel'>Relplot</a>
With this `flf` function we can plot two different `alf`: `scatteplot` and `lineplot`.

```python
sns.set_theme(style = 'darkgrid')
tips = sns.load_dataset("tips")
tips.head(6)
```

```python
g = sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time',
                kind = 'scatter', facet_kws = {'legend_out' : False})
g.figure.suptitle('Scatterplot', y =  1.05, fontsize = 15, style = 'oblique')
```

>**Note**: Not sure if in newer *Seaborn* versions, the `facet_kws` content can be pass as normal attributes and *Seaborn* manages it behind scenes.

```python
fmri = sns.load_dataset("fmri")
fmri.head(6)
```

```python
g = sns.relplot(data = fmri, x = 'timepoint', y = 'signal', col = 'region',
                kind = 'line', facet_kws = {'legend_out' : False})
g.figure.suptitle('LinePlot', y =  1.05, fontsize = 15, style = 'oblique')
```

By default, `lineplot` adds a 95% `ci` (confidence interval) along the line. This `ci` is representing *uncertainty*.

Of course, another way to plot a `scatterplot` and a `lineplot` would be by just calling their own functions:

```python
fig = plt.figure()
fig, ax = plt.subplots(nrows = 1, ncols = 2, layout = 'constrained', figsize = (8,4))
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time', ax = ax[0])
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'region',ax = ax[1])
```

>As I said in the *0.Seaborn_intro*, it is a good practice to use the `flf`(**relplot**) to plot an `alf` (**scatterplot and lineplot**) instead of using the `alf` directly and I justified why. <br><br>
That's why in this noteebook, you won't see anymore neither `scatterplot()` nor `lineplot()`, but `relplot` instead.

### <a id='scatter'>Scatterplot</a>

- This is the `alf` that `relplot` uses by default.


- The truth is that we've seen most of the `scatterplot` parameters so far. [Here](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) is the documentation.


- We are going to be seeing some examples, but most of the attributes should be familiar because they have  been mentioned in past notebooks.

You can plot up to  5 different variables in a *scatterplot* with the attributes `hue`, `style` and `size`.

```python
'''
Adding different color and style to the variable smoker
in order to make the plot more accessible.
''' 
# three variables are being plotted: total_bill, tip and smoker
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker")
```

```python
# now 4 variables are being plotted. The 3 above and  time
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
```

```python
# now 5 variables are being plotted. The 4 above and  sex
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", 
            style="time", size = 'sex', sizes = [10, 20])
```

At this point is  better to plot multiple plots calling `col` or `row` attribute to visualize better the different variables in the dataset.

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", 
            row="time", col = 'sex', facet_kws = dict(margin_titles = True))
```

You can customize the `style` attribute with the `markers` attribute. [Here](https://matplotlib.org/stable/api/markers_api.html) you can find all kinds of markers.

```python
m = {'Lunch':'p', 'Dinner':'+'}
sns.relplot(data=tips, x="total_bill", y="tip", hue = 'time', style="time",
           markers = {'Lunch':'D', 'Dinner':'s'})
```

You can use `marker` even if you don't use `style`.

>**Note**: Notice that if you don't use `style`, then the attribute is `marker` without an **s** at the end.

```python
sns.relplot(data=tips, x="total_bill", y="tip", marker = '+')
```

### <a id='line'>Lineplot</a>
When we want to plot a continous variable (like time data), `lineplot` is our function.

This `alf` has more attributes than the last one. We'll see the most important ones.

```python
# Now the style changes a littles bit since we plot lines and not dots
sns.relplot(data=fmri, x="timepoint", y="signal", col="region",
            hue="event", style="event", kind="line")
```

#### ci

if we want to get rid of the `ci`(confidence interval) we can simply set `ci = False`, or `errorbar = False` in newer *Seaborn* versions.

```python
sns.relplot(data=fmri, x="timepoint", y="signal", col="region",
            hue="event", style="event", kind="line", ci = False)
```

#### linewidth

```python
sns.relplot(data=fmri, x="timepoint", y="signal", col="region",
            hue="event", style="event", kind="line", ci = False,
           lw = 3)
```

####  markers & makersize
When assigning a `style` variable, `markers` can be used along with `markersize` so set the desired size if the default size doesn't convice you.

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue="event", kind = 'line',
             style="event", markers=True, ci = False, markersize = 10)
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", hue="subject",
             kind = 'line', style="event", ci = False)
```

#### units
When used, a separate line will be drawn for each variable value. 

Notice that every subject in the dataset is repeated and has `event` equal to **stim** and **cue**.

```python
fmri.loc[fmri['subject'] == 's13']
```

```python
# using units. Graphs all subject's values grouped by event
sns.relplot(data=fmri.query("region == 'frontal'"), x="timepoint", 
             kind = 'line',y="signal", units = 'subject', hue="event",
            ci = False,  estimator = None)
```

#### sort
By default, the observations are sorted by `x`. Disable this to plot a line with the order that observations appear in the dataset:

```python
x, y = np.random.normal(size=(2, 5000)).cumsum(axis=1)
sns.relplot(x=x, y=y, kind = 'line', sort=False, lw=1)
```

### <a id='rug'>rugplot</a>

Even though this is a `displot alf`, I want to mention this plot here because it is used with `scatterplot` (and `kdeplot` as we'll see in the next notebook) to take a better look at where is our data more concentrated. 

- In other words, with `rugplot()` we can see better where we have the most amount of data.

Rugplot draws marginal distributions by drawing ticks along the x and y axes.

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", kind = 'scatter')
sns.rugplot(data=tips, x="total_bill", y="tip", hue="time")
```
