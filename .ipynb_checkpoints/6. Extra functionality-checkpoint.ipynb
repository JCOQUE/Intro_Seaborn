## <a id= 'vs'>Figure-level functios vs Axis-level functions</a>

Remember when in the *0. Seaborn_intro* notebook I mentioned that *Matplotlib* is used for more controlled and customized plots and *Seaborn* for quicker and nicer plots?
- And after that, I said that since *Seaborn* is built on top of *Matplotlib*, it allows you to use it's API. Therefore, I'd use *Seaborn* (almost) always because you get the best of both worlds.

> A situation where we would use *Matplotlib* or a *Seaborn* `alf` instead of a *Seaborn* `flf` (as we've doing) is when you want **full flexibility in subplots** (as we saw in the *1. Subplots* notebook; or if our data **is not in a dataset** but in lists or numpy arrays instead.

We've been covering this throughout the whole tutorial. But this time, take deeper a look at the object these functions return, and the difference between them

```python
facetGrid_object = sns.relplot(data = tips, x = 'total_bill', y = 'tip')
type(facetGrid_object) #Facetgrid object
```

```python
axis_object = sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
type(axis_object) # axesSubplot object
```

The key difference is that with the first one, the `FacetGrid object` that we obtain when using a **figure-level function**, has the `ax` method that allows us to access the *Matplotlib axis* and use the *Matplotlib API*. 

> I like to see it as if with the `ax` method, you were transported to the *Matplotlib* world from the *Seaborn* world.

So, you can set any functionality that you want covered in the *Matplotlib tutorial*. Check out the last *Matplotlib* notebook in the * Plot and basic functionality*  folder called **Advanced Axis configuration** to get a better idea of different things you can modify with the `ax` method (set_xlabel, set_ylabel, xtikcs, set_title, etc.).

```python
# the plot of before
ticks = ['T', 'I', 'C', 'K', 'S']
x = np.arange(10, 60, 10)
facetGrid_object = sns.relplot(data = tips, x = 'total_bill', y = 'tip')
facetGrid_object.ax.set_title('New title')
facetGrid_object.ax.set_facecolor('k')
facetGrid_object.ax.set_xticks(ticks = x, labels = ticks)
facetGrid_object.ax.axline((0,0), slope = 0.15, color = 'yellow', lw = 2, ls = ':')
```

And not only the axis, you can access the figure configuration as well.

```python
facetGrid_object = sns.relplot(data = tips, x = 'total_bill', y = 'tip')
facetGrid_object.ax.set_title('New title')
facetGrid_object.ax.set_facecolor('k')
facetGrid_object.ax.axline((0,0), slope = 0.15, color = 'yellow', lw = 2, ls = ':')
facetGrid_object.figure.suptitle('Supertitle', y = 1.1)
facetGrid_object.figure.text(0.2, 0.9, 'This is my text', color = 'w')
facetGrid_object.figure.supxlabel('Super_X_label', y = -0.05)
facetGrid_object.figure.supylabel('Super_Y_label', x = -0.05)
```

The variable `a` does also have a dictionary of axis in case we had subplots. This allows us to iterate over all subplots

```python
facetGrid_object = sns.relplot(data = tips, x = 'total_bill', y = 'tip',
                hue="smoker", col = 'time', row = 'sex', height=4)
title_counter = 1
for ax in facetGrid_object.axes_dict.values():
    ax.set_title(f'Subplot {title_counter}')
    title_counter += 1
```

But don't forget that `facetGrid_object` is a `FacetGrid` object. As we saw in the *1. Subplots* notebook, *FacetGrid's object* also have their own methods. 

This increments even more the flexibility.

```python
facetGrid_object = sns.relplot(data = tips, x = 'total_bill', y = 'tip')
facetGrid_object.ax.set_title('New title')
facetGrid_object.ax.set_facecolor('k')
facetGrid_object.ax.axline((0,0), slope = 0.15, color = 'yellow', lw = 2, ls = ':')
facetGrid_object.despine(offset = 6) # facet grid method
```

But... what happens if I'm using a `alf` for plotting my data. We saw in the beginning of this notebook that the object returned was not a `FacetGrid` object.

Since the returned object is an `AxesSubplot` object, it is easier. We can access all the method seen before (except the `FacetGrid` ones) directly.

```python
g = sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
type(g)
```

```python
axis_object = sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
axis_object.set_title('New title')
axis_object.set_facecolor('k')
axis_object.axline((0,0), slope = 0.15, color = 'yellow', lw = 2, ls = ':')
```

And additionally, we can access with the `axis_object` variable to figure methods the same way we did with the `FacetGrid` object

```python
axis_object = sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
axis_object.set_title('New title')
axis_object.set_facecolor('k')
axis_object.axline((0,0), slope = 0.15, color = 'yellow', lw = 2, ls = ':')
axis_object.figure.suptitle('Supertitle', y = 1.1)
axis_object.figure.text(0.2, 0.9, 'This is my text', color = 'w')
axis_object.figure.supxlabel('Super_X_label', y = -0.05)
axis_object.figure.supylabel('Super_Y_label', x = -0.05)
```

Needless to say that you can use the *Matplotlib* API whenever you want.

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('figure title')
plt.xlim(5, 40)
plt.ylim(0, 8)

plt.show()
```

## <a id = 'color'>Color properties</a>
You can find [here](https://seaborn.pydata.org/tutorial/properties.html) more **style properties** but I'm going to focus on  the color.

*Seaborn* attributes such as `hue` use different color for nominal and numeric variables. 

<div>
    <img src = attachment:image.png>
</div>

For **nominal** variables, the colors are very distinct from one another.
On the other hand, for **numeric** variables, it's the same color that goes from ligther to darker.

Let's see an example

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', 
            hue = 'day', kind = 'scatter')
plt.title('Nominal variable')
```

```python
g = sns.relplot(data = tips, x = 'total_bill', y = 'tip',
            hue = 'size', kind = 'scatter')
g.figure.suptitle('Numeric variable')
```

Also there is the `alpha` variable, that we can use to modify the opacity of our colors in the plot.

```python
sns.catplot(data = tips, x = 'sex', y = 'total_bill', 
            kind = 'bar', alpha = 0.3, ci = None)
```
